```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-2, 0, 21)
y = x**2
```

```python
N = 20
y_fourier = 8/3 / 2
for n in range(1, N+1):
    a_n = 16/np.pi**2/n**2 * (-1)**n
    y_fourier += a_n*np.cos(n*x*np.pi/2)
```

```python
plt.plot(x, y, label='Origin')
plt.plot(x, y_fourier, label='Fourier')
plt.legend()
plt.show()
```
