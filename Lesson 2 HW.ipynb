#### За книгу заплатили 100р. и ещё половину своей стоимости. Сколько стоит книга?

Пусть х - стоимость книги, тогда 100 + х/2 = х, откуда х/2 = 100, х = 200.

#### За книгу заплатили 100р., и осталось заплатить ещё столько, сколько осталось бы заплатить, если бы за неё заплатили столько, сколько осталось заплатить. Сколько стоит книга?

Пусть х - стоимость книги, тогда осталось заплатить х - 100, по условию задачи:
    
    х - 100 = х - (х - 100)
    х - 100 = 100
    х = 200
